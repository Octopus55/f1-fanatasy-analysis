```python
class opt_val:
    def __init__(self, item, raw_val, cost, item_type):
        self.item = item
        self.raw_val = raw_val
        self.cost = cost
        self.type = item_type

    def __str__(self):
        return "{} {} {} {}".format(self.item, self.raw_val, self.cost, self.type)
```

```python
# create list of opt_val types to iterate through
opt_vals = dict()

for index, row in optimization_vals.iterrows():
    opt_vals[row['Item']] = opt_val(row['Item'], row['raw value'], row['Cost'], row['Type'])
```

```python
solved_solutions = pd.DataFrame(columns=['weight','num_teams','num_drivers','score'])

# solution will hold a list of dataframe tuples
def recur_team(weight=100, num_teams=2, num_drivers=5, Items=None, solution=0, solved_solutions=solved_solutions):
    if (num_teams==0) and (num_drivers==0):
        solved_solutions = pd.concat([solved_solutions, 
                                     pd.DataFrame([{'weight':weight,
                                                   'num_teams':num_teams,
                                                   'num_drivers':num_drivers,
                                                   'score':solution}])])
        return solution, solved_solutions
    elif (weight < 0 and ((num_drivers > 0) or (num_teams > 0))):
        return -1, solved_solutions
    
    # return a memoized solution
    if ((solved_solutions['weight']==weight) & 
        (solved_solutions['num_teams']==num_teams) & 
        (solved_solutions['num_drivers']==num_drivers)).any():
        return solved_solutions.loc[('weight'==weight) & ('num_teams'==num_teams) & ('num_drivers'==num_drivers), 'score'], solved_solutions

    max_solution = solution
    for item in opt_vals:
        if opt_vals[item].type == "Driver":
            partial_sol, solved_solutions = recur_team(weight=weight-opt_vals[item].cost, 
                                                       num_drivers=num_drivers-1,
                                                       num_teams=num_teams, 
                                                       solution=solution+opt_vals[item].raw_val)
        else:
            partial_sol, solved_solutions = recur_team(weight=weight-opt_vals[item].cost, 
                                                       num_drivers=num_drivers,
                                                       num_teams=num_teams-1,
                                                       solution=solution+opt_vals[item].raw_val)
        if partial_sol == -1: continue

        if partial_sol > max_solution: max_solution=partial_sol

    solved_solutions = pd.concat([solved_solutions, 
                                 pd.DataFrame([{'weight':weight,
                                               'num_teams':num_teams,
                                               'num_drivers':num_drivers,
                                               'score':solution}])])

    return max_solution, solved_solutions
        
    
```
